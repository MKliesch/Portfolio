# Predicting Cancer Using SVM and Decision Tree Classification


## Introduction

The goal of this project is to compare the performance of Decision Trees (DT) and Support Vector Machine (SVM) on a dataset using human cell records to classify cells in terms of whether samples are benign (mild state) or malignant (evil state).

The dataset consists of approximately 700 entries with each sample being characterized according to 10 different features (such as uniformity of cell size, uniformity of cell shape etc.). SVMs are known to perform particularly well in high-dimensional data and with small datasets. DTs, too, can reach very high accuracy if allowed an infinite number of levels. However, the higher the number of levels, the higher the risk is of overfitting and the more difficult it becomes to interpret the final DT. If pruned to an optimal state, DTs can often provide a more intuitive representation of the influence of given predictors, which, however, often comes at the cost of model accuracy.

To verify if this is the case, I will optimize a DT and a SVM for the cancer dataset and contrast five different performance measures against one another.

### Research Question

1. How accurately can benign and malignant cell samples be categorized based on human cell records?
2. Which machine learning algorithm yields a higher performance?

## Decision Tree

### Unspecified Decision Tree to Optimize

First, I will model a decision tree without specifying any optimization parameters. In a second step, I will determine the best fitting model through repeated Cost Complexity Pruning and cross-validation.

```python
# Create decision tree object
DT_final = DecisionTreeClassifier()

# Fit to training data
DT_final.fit(X_train, y_train)

# Predict testing set
yhat_basic = DT_final.predict(X_test)
yhat_basic[0:10]
```

```python
# Plot tree
plt.figure(figsize = (15,7))
plot_tree(DT_final,
         filled = True,
         rounded = True,
         class_names = ["Benign", "Malignant"],
         feature_names = data.columns[1:]);
```

The tree has a high number of levels, which can make it prone to overfitting and makes it difficult to interpret.

```python
# Check tree accuracy with confusion matrix
plot = plot_confusion_matrix(DT_final, X_test, y_test, display_labels = ["Benign", "Malignant"])
```

```python
# Compute confusion matrix
cnf_matrix = confusion_matrix(y_test, yhat_basic, labels=[2,4])
np.set_printoptions(precision=2)

# Compute accuracy measures
TP = cnf_matrix[1][1]
TN = cnf_matrix[0][0]
FP = cnf_matrix[0][1]
FN = cnf_matrix[1][0]

DT_final_accuracy = (TP + TN) / (TP + FP + FN + TN)
print("Accuracy of DT is ",DT_accuracy*100,"%")

DT_final_sensitivity = (TP) / (TP + FN)
print("Sensitivity of DT is ",DT_final_sensitivity*100,"%")

DT_final_specificity = (TN) / (FP + TN)
print("Specificity of DT is ",DT_final_specificity*100,"%")

DT_final_posPredVal = (TP) / (TP + FP)
print("Positive predictive value of DT is ",DT_final_posPredVal*100,"%")

DT_final_negPredVal = (TN) / (FN+TN)
print("Negative predictive value of DT is ",DT_final_negPredVal*100,"%")
```

Apart from 7 cases, the accuracy of this decision tree is fairly good. However, due to the high number of branches, it is difficult to interpret the classification. In addition, despite the high accuracy, there could still be a better model, which will be determined in the following section.

#### Optimizing the Decision Tree

Let's optimize the performance of the decision tree through ten-fold cross-validation to determine the best-performing value of alpha for complexity cost pruning.

```python
# increase number of decimals to avoid rounding to identical values
np.set_printoptions(formatter={'float': lambda x: "{0:0.4f}".format(x)})

# extract all possible values of alpha from the basic model
path = DT_final.cost_complexity_pruning_path(X_train, y_train) # determine values for alpha
ccp_alphas = path.ccp_alphas # extract different values for alpha
ccp_alphas = ccp_alphas[:-1] # exclude the maximum value for alpha
ccp_alphas
```

```python
## create an array to store the results of each fold during cross validation
alpha_loop_values = []

## For each candidate value for alpha, run 10-fold validation
## Store the mean and SD of the scores for each call
for ccp_alpha in ccp_alphas:
    dt = DecisionTreeClassifier(random_state = 4, ccp_alpha = ccp_alpha)
    scores = cross_val_score(dt, X_train, y_train, cv = 10)
    alpha_loop_values.append([ccp_alpha, np.mean(scores), np.std(scores)])

## Draw graph of means and sds of the scores for 
## each candidate value of alpha
alpha_results = pd.DataFrame(alpha_loop_values,
                            columns = ["alpha",
                                      "mean_accuracy",
                                      "std"])
alpha_results.plot(x='alpha',
                  y = 'mean_accuracy',
                  yerr = 'std',
                  marker = 'o',
                  linestyle = '--');
```

It seems that the 5th alpha value yields the highest accuracy when randomly selecting 10 different training/testing sets.

```python
# determine the 5th alpha value
best_alpha = float(ccp_alphas[4])
best_alpha
```

#### Final Decision Tree

```python
# Create decision tree object
DT_final = DecisionTreeClassifier(ccp_alpha = best_alpha)

# Fit to training data
DT_final.fit(X_train, y_train)

# Predict testing set
yhat = DT_final.predict(X_test)
yhat[0:10]
```

```python
# Plot tree
plt.figure(figsize = (15,7))
plot_tree(DT_final,
         filled = True,
         rounded = True,
         class_names = ["Benign", "Malignant"],
         feature_names = data.columns[1:]);
```

The pruned decision tree has been pruned to seven levels and is now easier to interpret but still quite big. According to this new tree, the most indicative parameters to classify benign from malignant cases are uniformity of cell size, uniformity of cell shape and the number of bare nuclei. Let's see how the pruned tree performs.

```python
# Check tree accuracy with confusion matrix
plot = plot_confusion_matrix(DT_final, X_test, y_test, display_labels = ["Benign", "Malignant"])
```

The new model has become more conservative, i.e. categorizing one more benign case as malignant (from 3 to 4) but reducing the number of false negatives from 4 to 2. Let's take a look at the new performance scores.

```python
# Compute confusion matrix
cnf_matrix = confusion_matrix(y_test, yhat, labels=[2,4])
np.set_printoptions(precision=2)

# Compute accuracy measures
TP = cnf_matrix[1][1]
TN = cnf_matrix[0][0]
FP = cnf_matrix[0][1]
FN = cnf_matrix[1][0]

DT_final_accuracy = (TP + TN) / (TP + FP + FN + TN)
print("Accuracy of final DT is ",DT_accuracy*100,"%")

DT_final_sensitivity = (TP) / (TP + FN)
print("Sensitivity of final DT is ",DT_final_sensitivity*100,"%")

DT_final_specificity = (TN) / (FP + TN)
print("Specificity of final DT is ",DT_final_specificity*100,"%")

DT_final_posPredVal = (TP) / (TP + FP)
print("Positive predictive value of final DT is ",DT_final_posPredVal*100,"%")

DT_final_negPredVal = (TN) / (FN+TN)
print("Negative predictive value of final DT is ",DT_final_negPredVal*100,"%")
```

### SVM

SVM works by mapping data to a high dimensional feature space, finding a (linear or non-linear) separator between categories and then transforming the data in such a way that the separator could be drawn as a hyperplane.

#### Default SVM Classifier to Optimize

```python
# SVM object using Support Vector Classifier

'''
kernel : string, optional (default='rbf')
 |      Specifies the kernel type to be used in the algorithm.
 |      It must be one of 'linear', 'poly', 'rbf', 'sigmoid', 'precomputed' or
 |      a callable.
 
gamma : {'scale', 'auto'} or float, optional (default='scale')
 |      Kernel coefficient for 'rbf', 'poly' and 'sigmoid'.
 |  
 |      - if ``gamma='scale'`` (default) is passed then it uses
 |        1 / (n_features * X.var()) as value of gamma,
 |      - if 'auto', uses 1 / n_features.
 
C is penalty that you want to impose (i.e. regularization parameter). 
 |      That is, how strictly the algorithm should handle datapoints that are incorrectly placed on the hyperplane. The strength of the regularization is
 |      inversely proportional to C. Must be strictly positive. The penalty
 |      is a squared l2 penalty.


'''

svmClassifier = svm.SVC(kernel = 'linear', gamma = 'auto', C=2)
svmClassifier.fit(X_train, y_train)
yhat_basic = svmClassifier.predict(X_test)
yhat_basic[0:5]
```

```python
# Check tree accuracy with confusion matrix
plot = plot_confusion_matrix(svmClassifier, X_test, y_test, display_labels = ["Benign", "Malignant"])
```

The SVM classifier did not identify any malignant cases as benign, but classified 5 benign cases as malignant and is hence even more conservative than the pruned decision tree. 

Let's see if we can optimize the SVM's performance to increase sensitivity.

```python
# Compute confusion matrix
cnf_matrix = confusion_matrix(y_test, yhat_basic, labels=[2,4])
np.set_printoptions(precision=2)

# Compute accuracy measures
TP = cnf_matrix[1][1]
TN = cnf_matrix[0][0]
FP = cnf_matrix[0][1]
FN = cnf_matrix[1][0]

SVM_basic_accuracy = (TP + TN) / (TP + FP + FN + TN)
print("Accuracy of basic SVM is ",SVM_basic_accuracy*100,"%")

SVM_basic_sensitivity = (TP) / (TP + FN)
print("Sensitivity of basic SVM is ",SVM_basic_sensitivity*100,"%")

SVM_basic_specificity = (TN) / (FP + TN)
print("Specificity of basic SVM is ",SVM_basic_specificity*100,"%")

SVM_basic_posPredVal = (TP) / (TP + FP)
print("Positive predictive value of basic SVM is ",SVM_basic_posPredVal*100,"%")

SVM_basic_negPredVal = (TN) / (FN+TN)
print("Negative predictive value of basic SVM is ",SVM_basic_negPredVal*100,"%")
```

#### Optimizing the SVM Classifier

Let's optimize the SVM classifier by adjusting the kernel and C (the regularization parameter) using `GridSearchCV`.

```python
parameters = {'kernel':('linear', 'rbf', 'poly'), 'C':[1,2,3]}
svmClassifier = svm.SVC()
clf = GridSearchCV(svmClassifier, parameters)
clf.fit(X_train, y_train)
```

We have now created several SVMs where every combination of the given parameters (3 kernels and 3 values for C) has been modelled and tested against one another. Let's see which kernel and regularization parameter yielded the best result:

```python
clf.best_params_
```

Let's rerun the final SVM by using these settings.

```python
svmClassifier = svm.SVC(kernel = 'rbf', gamma = 'auto', C=1)
svmClassifier.fit(X_train, y_train)
yhat_final = svmClassifier.predict(X_test)
yhat_final[0:5]
```

```python
# Check tree accuracy with confusion matrix
plot = plot_confusion_matrix(svmClassifier, X_test, y_test, display_labels = ["Benign", "Malignant"])
```

On our current split of training/test data, the new SVM performs exactly the same as the unspecified SVM. Hence, for final evaluation, we can compare the performance of the pruned decision tree with that of any of the two SVMs.

## Discussion

This analysis showed that both the DT and SVM yielded accuracy values of up to 96%, with SVM slightly outperforming DT. In addition, SVM performed better in terms of sensitivity (i.e. classifying malignant cases as malignant), but the DT was more specific(i.e. more consistently classifying benign samples as benign). This means that for this dataset, SVM resulted in a more conservative classification, which might classify more benign cases as malignant than the DT. In a real case scenario, where treatment decisions depend on the classification of cell samples, a more conservative approach might be preferred in order to prevent cancerous cells from being classified as harmless. 

However, the difference in performance between DT and SVM was only marginal, but the DT has the advantage of also providing a visual and illustrative approach that can compensate the slightly lower accuracy obtained. Visualization of the classification tree can help identify predictors that are more representative of a given class then others, which in turn can help optimize the number and type of predictors for future models.

In a real-case scenario, I would not only have optimized the models themselves but also the set of predictors so as to reduce the amount of noise in the dataset. 

In conclusion, for this dataset, both SVM and DT performed surprisingly well in identifying cancerous cell samples. A combination of SVM and DT visualization are the suggested methods to be included in the evaluation of malignant cell states.
